### 1. Random Quantum Circuit

```python
def circuit():
    qc = QuantumCircuit(4,4)
    qc.h(0)
    qc.h(2)
    qc.u3(pi/2,pi/2,pi/2,0)
    qc.cu1(pi/2,0, 1)
    qc.u2(pi/2,pi/2,1)
    qc.u1(pi/2,2)
    qc.x(0)
    qc.ccx(1,2,3)
    qc.y(1)
    qc.z(2)
    qc.cx(2,3)
    qc.z(3)
    qc.h(3)
    qc.s(0)
    qc.cu1(pi/2,1, 2)
    qc.swap(0,2)
    qc.cswap(0,1,3)
    return qc
```

#### 1.1 Circuit diagram

```python
qc = circuit()
style = {'backgroundcolor': 'lavender'}
qc.draw(output='mpl', style = style)
```

#### 1.2 Two qubit system (state vector)

```python
qc = circuit()
writeState(qc)
```

 ##### 3.3.1 Phase

```python
def getPhaseSeq():
    phaseDic = []
    
    qc = QuantumCircuit(4,4)
    qc0 = qc.copy()
    
    qc.h(0)
    qc.h(2)
    qc.h(3)
    qc1 = qc.copy()
    
    qc.u3(pi/2,pi/2,pi/2,0)
    qc2 = qc.copy()
    
    qc.cu1(pi/2,0, 1)
    qc.u2(pi/2,pi/2,1)
    qc3 = qc.copy()
    
    qc.u1(pi/2,2)
    qc.x(0)
    qc.ccx(1,2,3)
    qc4 = qc.copy()
    
    qc.y(1)
    qc.z(2)
    qc.cx(2,3)
    qc5 = qc.copy()
    
    qc.z(3)
    qc.h(3)
    qc.s(0)
    qc6 = qc.copy()
    
    qc.cu1(pi/2,1, 2)
    qc.swap(0,2)
    qc.cswap(0,1,3)
    qc7 = qc.copy()
    
    
    for iqc in [qc0,qc1,qc2,qc3,qc4,qc6,qc7]:
        phaseDic.append(getPhase(iqc))
    return phaseDic    
```

```python
fsize = [1.5*6,1.8*12]
drawPhase(getPhaseSeq(),fsize)
```
